```python
CRT = [['.'] * 40 for _ in range(6)]

spritePosition = 1  # +-1
currCycle = 1  # draw to position -1
remCycles = 0
multiplyCycle = 20

signalStrengths = list()

while len(input) > 0:
    # load instruction
    if remCycles < 1:
        currInstruction = input.pop(0)
        if currInstruction.split()[0] == 'addx':
            remCycles = 2
    
    # during the cycle
    if currCycle == multiplyCycle:
        signalStrengths.append(spritePosition * multiplyCycle)
        multiplyCycle += 40
    
    # draw CRT
    crt_drawTo = currCycle - 1
    crt_linePosition = crt_drawTo % 40
    if crt_linePosition in [spritePosition - 1, spritePosition, spritePosition + 1]:
        crt_line = crt_drawTo // 40
        CRT[crt_line][crt_linePosition] = '#'
    
    # set instruction
    currCycle += 1
    remCycles -= 1
    if remCycles < 1:
        if currInstruction.split()[0] == 'addx':
            spritePosition += int(currInstruction.split()[1])

CRT = [''.join(x) for x in CRT]
```
